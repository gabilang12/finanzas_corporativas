# **Web-Scrapin**

Lo que se quiere realizar es traer todos los datos que aprecen en el URL de cada cuenta financiera tal cual aparecen en la pagina de yahoo finance, por esto se va a realizar web-scraping que consiste en recorrer toda la pagina proporcionada, buscando la informacion que se requiere y guardarla en un dataframe para luego poderla usar para realizar calculos (toda la informacion va a venir como un string)

```python
# Enter a stock symbol
index= 'FDX'
# URL link 
url_is = 'https://finance.yahoo.com/quote/' + index + '/financials?p=' + index
url_bs = 'https://finance.yahoo.com/quote/' + index + '/balance-sheet?p=' + index
url_cf = 'https://finance.yahoo.com/quote/' + index + '/cash-flow?p=' + index
```

```python
def findAndReplace(lista, itemAReemplazar, nuevoValor): 
    for index, item in enumerate(lista):
        if item == itemAReemplazar:
            lista[index] = nuevoValor
```

## **Income Statements**

```python
#si quiere revise que el url si funciona
url_is
```

```python
import requests
read_data = requests.get(url_is,headers={'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'})
```

```python
soup_is= BeautifulSoup(read_data.content,"lxml")
```

```python
divTag = soup_is.find_all("div", {"class": {"D(tbc)", "Ta(c)"}})

ls= []
ls.append("Breakdown")
```

```python
for tag in divTag:
  tag_value =  tag.get('value')
  ls.append(tag.text)

ls
```

```python
#ASEGURARSE DE QUE TODOS LOS '-' SI SE HALLAN REEMPLAZADO
for i in ls:
  if i == '-': findAndReplace(ls, '-', '0')
  print(i)
```

```python
#eliminamos toda la informacion que se extrajo de la pagina que esta entre 'breakdown' y 'ttm', ya que es informacion que si bien tiene la misma clase 
#con la cual vamos a extraer la informacion del estado de resultados, es informacion que no debe ir ahi (anuncios, espacios en blanco, etc.)
contador = 0
for i in ls:
  if i=='ttm':
    del ls[1:contador]
  else: contador += 1

ls
```

```python
#en la pagina la tabla tiene 6 columnas, por lo que vamos a separar los datos de la lista en 6, creando de esta forma una matriz de 6 columnas y n filas
new_ls = list(filter(None,ls))
is_data = list(zip(*[iter(new_ls)]*6))
is_data
```

```python
#creamos el Dataframe
Income_st = pd.DataFrame(is_data[0:])
Income_st
```

```python
Income_st.columns = Income_st.iloc[0] #Nombramos las columnas con la primera fila del dataframe
Income_st.drop(Income_st.index[0],inplace=True) #eliminamos la primera fila
Income_st
```

```python
#revisamos si hay una ultima fila que tenga informacion que no corresponde con la informacion financiera y la eliminamos
fila = 34 #numero de la fila que hay que eliminar
Income_st = Income_st.drop(fila,axis=0)
Income_st
```
