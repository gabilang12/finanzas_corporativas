## Medias móviles (simples y exponenciales)

```python
data = pd.DataFrame()
data['Close'] = df['Close']
```

### Media Móvil Simple

La media móvil o simple moving average (SMA) en finanzas, es un cálculo aritmético que divide la suma de los precios (generalmente de cierre) para un determinado periodo dividido entre el número de periodos utilizado. La media móvil puede ascender, mantenerse o descender en base a si el nuevo precio es superior, igual o inferior a la media en el periodo anterior.

```python
data['SMA'] = data.iloc[:,0].rolling(window=50).mean() #windows es la cantidad de periodos
data
```

```python
data.plot(kind='line')
```

### Media Móvil Exponencial

La media móvil exponencial o EMA (Exponential Moving Average) le da más peso a periodos actuales. Se asigna un factor multiplicador que crece de forma inversa al número de periodos.

```python
data['EMA'] = data.iloc[:,0].ewm(span=50).mean()
data
```

```python
data.plot(kind='line')
```

## Bandas de Bollinger

Las bandas de bollinger delimitan una zona determinada por una media móvil en el centro de 20 periodos a partir del cálculo de la desviación estándar, que es una forma de entender la volatilidad de un activo o cartera. 2 desviaciones estándar, que son las que se suelen usar para este indicador, son aproximadamente un 95,4% de los casos, basándonos en un sistema de distribución normal de la varianza del precio.

Si el activo se sale de estan bandas puede ser una aituación para comprar o vender.

```python
N = 20 #periodos de la media movil simple
k = 2  #factor para las bandas (desviaciones)

data['Standard Deviation'] = data['Close'].rolling(N).std()
data['Middle Band'] = data['Close'].rolling(N).mean()
data['Upper Band'] = data['Middle Band'] + data['Standard Deviation'] * k
data['Lower Band'] = data['Middle Band'] - data['Standard Deviation'] * k

data
```

```python
plt.title("FedEX")

data['Close'].plot(x="Major", y="FedEX")
data['Upper Band'].plot(x="Major", y="", linewidth=0.5, color="#033660")
data['Lower Band'].plot(x="Major", y="", linewidth=0.5, color="#033660")
data['Middle Band'].plot(x="Major", y="", linewidth=0.5, color="#033660", linestyle="dashed")
```

## Indicadores RSI y MACD

### RSI

Este indicador, mide la velocidad y el cambio de los movimientos de precios durante periodos determinados, generalmente 14. Es una división entre la media del sumatorio las veces en las que el precio ha aumentado y en las que ha disminuido durante el periodo escogido.

Si las bandas reciben algún tipo de corte, significa una situación de sobre compra (en caso de la banda superior) o de sobre venta (en caso de la banda inferior)

```python
retornos = pd.DataFrame()
retornos = df['Close']
returns = retornos.pct_change()[1:]
returns
```

```python
long_window = 14
up = returns.clip(lower=0)
down = -1*returns.clip(upper=0)

ema_up = up.ewm(span=long_window).mean()
ema_down = down.ewm(span=long_window).mean()

rs = ema_up/ema_down
rsi = 100-(100/(1+rs))
```

```python
signal = pd.DataFrame(index = df.index)
signal['RSI'] = rsi
df = df.iloc[long_window:]
signal = signal.iloc[long_window:]
```

```python
import matplotlib.gridspec as gridspec
grafico = plt.figure(figsize=(20,10))
tabla = gridspec.GridSpec(nrows=2, ncols=1, figure=grafico, height_ratios=[3,1])

graf_sup = plt.subplot(tabla[0,0])
graf_inf = plt.subplot(tabla[1,0])

graf_sup.plot(retornos, label='CLose')
graf_sup.set_title("Precio")

graf_inf.plot(signal['RSI'], 'b', label="RSI")
graf_inf.axhline(70, color='r', linestyle='--')
graf_inf.axhline(30, color='r', linestyle='--')
graf_inf.set_title("RSI")

plt.show()
```

### MACD

Es un método que compara dos medias móviles suavizadas exponenciales. Consta de 2 lineas, la roja (línea de activación, la cual es una media móvil exponencial de 9 períodos) y la azul (MACD, la cual es la diferencia entre 2 medias móviles exponenciales de 12 y 25 días).

Se puede analizar de 2 formas:
* Con las líneas: cuando la línea de activación cruza la del MACD es porque la acción va a ir al alza (la línea de activación cruza hacia abajo), si la cruza hacia arriba es porque la acción va a ir a la baja. Si se igualan las líneas significa una tendencia lateral
* Con las velitas: si están hacia arriba significa un movimiento al alza; si están hacia abajo, lo contrario

```python
retornos = pd.DataFrame()
retornos = df['Close']
retornos
```

```python
m_rap = retornos.ewm(span=12, adjust=False).mean()
m_lenta = retornos.ewm(span=26, adjust=False).mean()

macd = m_rap - m_lenta
ema9 = macd.ewm(span=9, adjust=False).mean()
histograma = macd - ema9
```

```python
signal = pd.DataFrame(index=retornos.index)
signal['signal'] = np.where(macd > ema9, 1, 0)
signal['position'] = signal['signal'].diff()
```

```python
import matplotlib.gridspec as gridspec
grafico = plt.figure(figsize=(20,10))
tabla = gridspec.GridSpec(nrows=2, ncols=1, figure=grafico, height_ratios=[3,1])

graf_sup = plt.subplot(tabla[0,0])
graf_inf = plt.subplot(tabla[1,0])

graf_sup.plot(retornos, label='CLose')
graf_sup.set_title("Precio")

graf_inf.plot(retornos.index, macd, 'b', label="MACD")
graf_inf.plot(retornos.index, ema9, 'r--', label="Signal")
graf_inf.bar(retornos.index, histograma, color=(histograma>0).map({True:'g', False:'r'}))
graf_inf.set_title("MACD")

plt.show()
```
