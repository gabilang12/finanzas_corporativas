## *Análisis Técnico*

Se trata de identificar patrones de comportamiento en la historia de un activo financiero con el fin de predecir lo que puede ocurrir con el precio.

```python
!pip install yfinance
```

```python
import pandas as pd
import yfinance as yf
from datetime import datetime
import mplfinance as mpf
```

```python
index = yf.Ticker("FDX") 

inicio = datetime(2021, 1, 1)
fin = datetime(2022, 10, 31)

df = index.history(start = inicio, end = fin)
df #nos interesa la columna 'close
```

## Gráfico de la acciones

```python
!pip install mplfinance
```

```python
mpf.plot(df, type = 'line')
```

```python
#Si quiero graficar mas de 2 acciones
#importamos otra accion en el mismo lapso de tiempo que la primera
inicio = datetime(2021, 1, 1)
fin = datetime(2022, 10, 31)

index = yf.Ticker("MSFT") 

df2 = index.history(start = inicio, end = fin)
df2
```

```python
#agregamos las columnas necesarias para que solo nos queden los valores de clausura de cada accion
df3 = pd.DataFrame()
df3['FDX'] = df['Close']
df3['MSFT'] = df2['Close']
df3.head()
```

```python
df3.plot(kind='line')
```
